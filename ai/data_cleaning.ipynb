```python
import pandas as pd
import numpy as np
```

```python
# load the data RISdata_rhopalosiphum_padi_1990_2020_weekly.csv
df = pd.read_csv('RISdata_rhopalosiphum_padi_1990_2020_weekly.csv')
df.head()
```

```python
df['date'] = pd.to_datetime(df['Year'].astype(str) + '-W' + df['Week Number'].astype(str).str.zfill(2) + '-1', format='%Y-W%U-%w')
```

```python
# Extract month and year
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year
```

```python
# Group by relevant columns and sum the totals
monthly_data = df.groupby(['Trap', 'Insect', 'year', 'month']).agg({
    'Total': 'sum',
    'Trap Operating Days': 'sum'
}).reset_index()
```

```python
monthly_data.head()
```

```python
monthly_data = monthly_data.sort_values(['year', 'month'])
```

```python
monthly_data.head()
```

```python
# export the data to a csv file
monthly_data.to_csv('RISdata_rhopalosiphum_padi_1990_2020_monthly.csv', index=False)
```
