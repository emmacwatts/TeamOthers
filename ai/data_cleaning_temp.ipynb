```python
import pandas as pd
import numpy as np
```

```python
# load the data and remove first 3 rows
df = pd.read_excel('./montly_rain/01-data-rain-1853-2024.xlsx', sheet_name='DATA', skiprows=3)
df.tail(50)
```

```python
# remove column : Total Annual Rain	, corrected 2004-2020
df = df.drop(columns=['Total Annual Rain', 'corrected 2004-2020'])
```

```python
df.head()
```

```python
# Melt the dataframe to convert months from columns to rows
df_melted = pd.melt(df, 
                    id_vars=['Year'], 
                    value_vars=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
                    var_name='month', 
                    value_name='monthly_rainfall')

# Rename the year column
df_melted = df_melted.rename(columns={'Year': 'year'})

# Convert month abbreviations to numbers
month_map = {'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6,
             'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12}
df_melted['month'] = df_melted['month'].map(month_map)

# Reorder columns
df_final = df_melted[['year', 'month', 'monthly_rainfall']]
```

```python
df_final.tail()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a pivot table to get years as rows and months as columns
pivot_df = df_final.pivot(index='year', columns='month', values='monthly_rainfall')

# Create a mask for missing data (True where data is missing)
missing_mask = pivot_df.isnull()

# Create the visualization
plt.figure(figsize=(15, 10))

# Create heatmap: Green for data, Red for missing
# Convert boolean mask to colors: False (has data) = 0 (green), True (missing) = 1 (red)
colors = missing_mask.astype(int)

plt.imshow(colors, cmap='RdYlGn_r', aspect='auto')

plt.title('Data Availability: Green = Data Present, Red = Missing Data', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Year', fontsize=14)
plt.xticks(range(12), range(1, 13))
plt.yticks(range(len(pivot_df.index)), pivot_df.index)

# Add colorbar
cbar = plt.colorbar()
cbar.set_ticks([0, 1])
cbar.set_ticklabels(['Data Present', 'Missing'])

plt.tight_layout()
plt.show()

# Print summary
print(f"Total missing values: {missing_mask.sum().sum()}")
print(f"Total data points: {pivot_df.size}")
print(f"Missing percentage: {(missing_mask.sum().sum() / pivot_df.size * 100):.1f}%")
```

```python
# export to df_final.csv
df_final.to_csv('monthly_rain_1853_2024.csv', index=False)
```
