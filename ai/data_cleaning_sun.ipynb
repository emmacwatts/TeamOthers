```python
import pandas as pd
import numpy as np
```

```python
# load the data RISdata_rhopalosiphum_padi_1990_2020_weekly.csv and sheet name is Data - mthly mean average temp skip the first row
df = pd.read_excel('./sun_montly/01-data-sun-1891-2024.xlsx', sheet_name='DATA', skiprows=2)
df.head()
```

```python
# drop "Annual Total Sunshine Hours"	and "Mean"
df = df.drop(columns=['Annual Total Sunshine Hours', 'Mean'])

```

```python
df.head()
```

```python
import pandas as pd

# Assuming your dataframe is called df
df_clean = df.melt(id_vars=['Year'], 
                   var_name='month', 
                   value_name='sun_hr')

# Convert month names to numbers
month_map = {
    'January': 1, 'February': 2, 'March': 3, 'April': 4,
    'May': 5, 'June': 6, 'July': 7, 'August': 8,
    'September': 9, 'October': 10, 'November': 11, 'December': 12
}

df_clean['month'] = df_clean['month'].map(month_map)
df_clean = df_clean.dropna()
df_clean['Year'] = df_clean['Year'].astype(int)
df_clean = df_clean.sort_values(['Year', 'month']).reset_index(drop=True)
```

```python
df_clean.head()
```

```python
# export the csv file
df_clean.to_csv('./sun_montly/sun_montly_cleaned.csv', index=False)
```
