```python
import pandas as pd
import numpy as np
```

```python
# load the data RISdata_rhopalosiphum_padi_1990_2020_weekly.csv and sheet name is Data - mthly mean average temp skip the first row
df = pd.read_excel('./montly_air_temp/01-data-1878-2023.xlsx', sheet_name='Data - mthly mean average temp', skiprows=1)
df.head()
```

```python
# Melt the dataframe to convert months from columns to rows
df_melted = pd.melt(df, 
                    id_vars=['Unnamed: 0'], 
                    value_vars=['January', 'February', 'March', 'April', 'May', 'June',
                               'July', 'August', 'September', 'October', 'November', 'December'],
                    var_name='month', 
                    value_name='monthly_air_temp')

# Rename the year column
df_melted = df_melted.rename(columns={'Unnamed: 0': 'year'})

# Convert month names to numbers
month_map = {'January': 1, 'February': 2, 'March': 3, 'April': 4, 'May': 5, 'June': 6,
             'July': 7, 'August': 8, 'September': 9, 'October': 10, 'November': 11, 'December': 12}
df_melted['month'] = df_melted['month'].map(month_map)

# Reorder columns
df_final = df_melted[['year', 'month', 'monthly_air_temp']]
```

```python
# export the csv
df_final.to_csv('monthly_air_temp_1878_2023.csv', index=False)
```
